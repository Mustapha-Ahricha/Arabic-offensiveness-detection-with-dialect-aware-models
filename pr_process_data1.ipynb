```python
import pandas as pd
```

```python
data1 = pd.read_csv('/content/preprocessed_dialect_tweets.csv')
```

```python
# Renommer les valeurs dans une colonne spécifique

data1['dialect'] = data1['dialect'].replace({1 : 'IQ', 2 : 'LY', 3 : 'QA', 4 : 'PL',5 : 'SY', 6 : 'TN', 7 : 'JO', 8 : 'MA', 9 : 'SA', 10 : 'YE', 11 : 'DZ', 12 : 'EG', 13 : 'LB', 14 : 'KW', 15 : 'OM', 16 : 'SD', 17 : 'AE', 18 : 'BH'})

# Afficher les premières lignes de la dataset après les modifications
print("\nAprès renommage des valeurs :")
print(data1.head())
```

```python
# Enregistrer le DataFrame combiné dans un nouveau fichier CSV
fichier_combine = 'Preprocessed_arabic_dialect.csv'
data1.to_csv(fichier_combine, index=False)
```
